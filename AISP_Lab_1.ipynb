```python
import librosa as lb
import matplotlib.pyplot as plt
import IPython.display as ipd
from IPython.display import Audio

y, sr = lb.load('Recording.wav')
lb.display.waveshow(y)
plt.title("speech signal waveform")
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.show()
```

```python
duration = lb.get_duration(y=y, sr=sr)
print(f"Duration of the signal: {duration} seconds")
magnitude_range = max(abs(y))
print(f"Magnitude range of the signal: {magnitude_range}")
```

```python
start_time_a3 = 0
end_time_a3 = 1
segment_a = y[int(start_time_a3 * sr):int(end_time_a3 * sr)]
lb.display.waveshow(segment_a)
plt.title("speech segment 1 waveform")
plt.ylabel("Amplitude")
Audio(data=segment_a,rate=sr)
```
