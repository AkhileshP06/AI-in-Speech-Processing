```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
y,sr = librosa.load(r"C:\Users\Admin\Downloads\Recording.wav")
plt.figure(figsize=(10,4))
plt.plot(y)
```

```python
fft_ = np.fft.fft(y)
print(fft_)
```

```python
plt.figure(figsize=(10,4))
plt.plot(abs(fft_))
# magnitudes of sinusoidal components of the wave.
```

```python
ifft_ = np.fft.ifft(fft_)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(ifft_)
plt.title('IFFT')
plt.subplot(1,2,2)
plt.plot(y)
plt.title(' original signal')
```
